The string features need to be encoded so that they can fed into the model

```python
#storing the names of the object type columns 
object_columns = train_data.select_dtypes(include=['object']).columns.tolist()

print(object_columns)
```

```python
#using the ordinal encoder to store all the string values
import sklearn
from sklearn.preprocessing import OrdinalEncoder
ord= OrdinalEncoder()

train_data[object_columns]= ord.fit_transform(train_data[object_columns])
test_data[object_columns] = ord.transform(test_data[object_columns])
```

Using Keras Tuner to tune hyperparameters for the model 

```python
import keras
from keras import layers, Sequential
def build_model(hp):
    model = keras.Sequential()
    layer1_units = hp.Int('layer1', min_value=2, max_value=32, step=2)
    print(f'Layer 1 units: {layer1_units}')
    model.add(layers.Dense(units=layer1_units, input_dim=45))

    num_layers = hp.Int('num_layers', 2, 20)
    print(f'Number of layers: {num_layers}')
    for i in range(num_layers):
        units = hp.Int('units_' + str(i), min_value=2, max_value=32, step=2)
        regularizer_type = hp.Choice('regularizer_type_' + str(i), ['l1', 'l2', 'l1_l2', 'none'])
        print(f'Layer {i+1} units: {units}, regularizer: {regularizer_type}')
        if regularizer_type == 'l1':
            regularizer = keras.regularizers.l1(hp.Float('l1_' + str(i), min_value=0.01, max_value=0.1, step=0.01))
        elif regularizer_type == 'l2':
            regularizer = keras.regularizers.l2(hp.Float('l2_' + str(i), min_value=0.01, max_value=0.1, step=0.01))
        elif regularizer_type == 'l1_l2':
            regularizer = keras.regularizers.l1_l2(l1=hp.Float('l1_' + str(i), min_value=0.01, max_value=0.1, step=0.01),
                                                   l2=hp.Float('l2_' + str(i), min_value=0.01, max_value=0.1, step=0.01))
        else:
            regularizer = None

        model.add(layers.Dense(units=units,
                               activation='relu',
                               kernel_regularizer=regularizer))

    model.add(layers.Dense(1, activation='linear'))
    optimizer = keras.optimizers.Adam(hp.Choice('lrate', [1e-4, 1e-3, 1e-2]))
    model.compile(optimizer=optimizer, loss='mean_absolute_error', metrics=['mean_absolute_error'])
    return model
```

```python
from keras_tuner import RandomSearch

tuner = RandomSearch(build_model, objective='val_mean_absolute_error',
                     max_trials=3,
                     directory='cross_val_dir',  # Use a new directory
                     project_name='proj1')
```

```python
tuner.search_space_summary()

# Perform hyperparameter search
tuner.search(x=x_train, y=y_train,
             validation_data=(x_val, y_val),
             epochs=10)
tuner.search_space_summary()

# Perform hyperparameter search
tuner.search(x=x_train, y=y_train,
             validation_data=(x_val, y_val),
             epochs=10)
```

```python
# Get the best model
best_model = tuner.get_best_models()[0]

# Print the summary of the best model
best_model.summary()
```

```python
history = best_model.fit(x_train, y_train,
                         validation_data=(x_val, y_val),
                         epochs=1000,
                         batch_size=32)
```
